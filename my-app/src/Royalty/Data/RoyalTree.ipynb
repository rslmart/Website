```python
import pickle
import graphviz
```

```python
with open('royaltree.pickle', 'rb') as f:
    tree = pickle.load(f)
print(len(tree))
tree
```

```python
# Create a new Graphviz graph
graph = graphviz.Digraph()

for person in tree.values():
    graph.node(person['id'])
for person in tree.values():
    if 'father' in person and person['father']:
        graph.edge(person['father'], person['id'])
    if 'mother' in person and person['mother']:
        graph.edge(person['mother'], person['id'])
    if 'issueList' in person and person['issueList']:
        for child_id in person['issueList']:
            graph.edge(person['id'], child_id)

# Render the graph to a file
graph.render('family_tree', format='png', view=True)
```
