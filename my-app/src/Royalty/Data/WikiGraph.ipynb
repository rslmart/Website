```python
import requests
from bs4 import BeautifulSoup

def fetch_wikidata(params):
    url = 'https://www.wikidata.org/w/api.php'
    try:
        return requests.get(url, params=params)
    except:
        return 'There was and error'
```

```python
response = requests.get("https://www.wikidata.org/wiki/Q3044", headers={'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/90.0.4430.212 Safari/537.36'})
soup = BeautifulSoup(response.content, 'html.parser')
print(soup)
```

```python
property_map = {}
for div in soup.find_all('div', attrs={'class': "wikibase-statementgroupview-property-label"}):
	a = div.find('a')
	if a.text == 'described by source':
		break
	property_map[a.get('href').split(':')[1]] = a.text
property_map
```

```python
id = 'Q3044'
# Create parameters
params = {
            'action': 'wbgetentities',
            'ids': id,
            'format': 'json',
            'languages': 'en'
        }
# fetch the API
data = fetch_wikidata(params)
# Show response
data = data.json()
data
```

```python
person = {}
person['label'] = data['entities'][id]['labels']['en']['value']
person['descriptions'] = data['entities'][id]['descriptions']['en']['value']
for p in property_map.keys():
	person[property_map[p]] = []
	for e in data['entities'][id]['claims'][p]:
		try:
			# Need special handling for
			# 'noble title' to gather 'start time:P580', 'end time:P582', 'followed by:P156', 'follows:P155'
			# 'spouse' to gather 'start time:P580', 'end time:P582'
			# 'unmarried partner' to gather 'start time:P580'
			# 'position held' to gather 'start time:P580', 'end time:P582', 'replace:P1365', 'replaced by:P1366'
			person[property_map[p]].append(e['mainsnak']['datavalue']['value'])
		except:
			print(e)
```
